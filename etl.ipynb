### Dimension date

```python
def create_date_table(start='2021-01-01', end='2025-12-22'):
    df = pd.DataFrame({'date': pd.date_range(start, end)})
    df['date_id'] = df.index + 1
    df['year'] = df.date.dt.year
    df['month'] = df.date.dt.month
    df['day'] = df.date.dt.day
    df['day_name'] = df.date.dt.day_name()  # Cambiado a day_name()
    df['day_week'] = df.date.dt.dayofweek
    df['week'] = df.date.dt.isocalendar().week  # Modificado a isocalendar().week
    df['quarter'] = df.date.dt.quarter
    
    df = df[['date_id', 'date', 'year', 'month', 'day', 'day_name', 'day_week', 'week', 'quarter']] 
    
    return df

```

```python
date_dim = create_date_table()
```
